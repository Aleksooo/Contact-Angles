```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def LJ(r, C6=1, C12=1):
    return C6/r**12 - C12/r**6
```

```python
def Buck(r, A=1, B=1, C=1):
    return A*np.exp(-B*r) - C/r**6
```

```python
plt.grid()
r = np.linspace(0.9, 2, 100)
plt.plot(r, LJ(r))
```

```python
plt.grid()
r = np.linspace(0.5, 2, 100)
plt.plot(r, Buck(r))
```
